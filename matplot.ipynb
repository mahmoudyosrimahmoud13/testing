# Matplotlib tutorial

### libraries

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

#### scatter plot

```python
x = np.random.random(50)*100
y = np.random.random(50)*100
#c = color , s = size 

plt.scatter(x,y,c='red',marker='*',s= 150,alpha= 0.3)
```

### line plot

```python
years = [2003 + x for x in range(20)]

whieghts = [20 + x for x in range(20)]

#plt.plot(years , whieghts, c= 'g',lw=2 , linestyle ='--')
plt.plot(years , whieghts, 'r--',lw=2 )

```

### barchart

```python
lang = ['pyhthon','Go','C#','Java']
votes = np.random.randn(4)
plt.bar(lang,votes,align='edge',width=0.5,edgecolor='g',lw = 7)
```

### histogram

```python
#mean = 20 , deviation = 1.5 , 
ages = np.random.normal(20,7.5,1000)
# plt.hist(ages,bins = 20)
plt.hist(ages,bins = 20,cumulative=True)
```

### piecharts


```python
votes = np.random.normal(80,9,len(lang))

plt.pie(votes,labels=lang,explode=[0,0.3,0,0],autopct='%0.2f%%',pctdistance=1.3,startangle=90)
```
